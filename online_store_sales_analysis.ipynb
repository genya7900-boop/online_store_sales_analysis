```python
# Import pandas library
import pandas as pd
import os

# Create charts directory
charts_dir = r"C:\Users\PC\DA portfolio\1\online_store_sales_analysis\charts"
os.makedirs(charts_dir, exist_ok=True)
print(f"Charts will be saved to: {charts_dir}")

# Define the file path
file_path = r"C:\Users\PC\DA portfolio\1\online_store_sales.csv"

# Load the CSV file into a pandas DataFrame
sales_data = pd.read_csv(file_path)
# Create a working DataFrame named df used by plotting cells
df = sales_data.copy()
# Ensure 'month' is datetime for plotting
if not pd.api.types.is_datetime64_any_dtype(df['month']):
    df['month'] = pd.to_datetime(df['month'])

# Display the first 5 rows to verify the data loaded correctly
print("First 5 rows of the dataset:")
display(sales_data.head())

# Get basic information about the dataset
print("\nBasic information about the dataset:")
print(f"Number of rows and columns: {sales_data.shape}")
print(f"Column names: {sales_data.columns.tolist()}")
print("\nData types:")
display(sales_data.dtypes)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the revenue line
ax.plot(df['month'], df['revenue'], marker='o', color='steelblue', linewidth=2, label='Revenue')

# Format the x-axis to show month and year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Add labels and title
plt.title('Monthly Revenue', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add data labels on each point
for date, revenue in zip(df['month'], df['revenue']):
    ax.text(date, revenue, f'${revenue:,}', 
            ha='center', va='bottom', fontsize=9, fontweight='bold',
            bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=2))

# Adjust layout
plt.tight_layout()

# Save the plot as PNG
plt.savefig(os.path.join(charts_dir, 'monthly_revenue.png'), dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
import os
print(os.getcwd())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Calculate average cart value
df['avg_cart_value'] = df['revenue'] / df['orders']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the average cart value line
ax.plot(df['month'], df['avg_cart_value'], marker='o', color='crimson', 
        linewidth=2, label='Average Cart Value')

# Format the x-axis to show month and year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Add labels and title
plt.title('Monthly Average Cart Value', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Cart Value ($)', fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add data labels on each point
for date, avg_value in zip(df['month'], df['avg_cart_value']):
    ax.text(date, avg_value, f'${avg_value:.2f}', 
            ha='center', va='bottom', fontsize=9, fontweight='bold',
            bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=2))

# Adjust layout
plt.tight_layout()

# Save the plot as PNG
plt.savefig(os.path.join(charts_dir, 'monthly_avg_cart_value.png'), dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate averages
avg_visitors = df['visitors'].mean()
avg_orders = df['orders'].mean()

# Calculate the conversion percentage
conversion_percentage = (avg_orders / avg_visitors) * 100
non_conversion_percentage = 100 - conversion_percentage

# Create data for the pie chart
labels = ['Visits Resulted in Orders', 'No Order']
sizes = [conversion_percentage, non_conversion_percentage]
colors = ['#4CAF50', '#C0C0C0']  # Green for conversions, light gray for non-conversions
explode = (0.1, 0)  # Explode the first slice (conversions) for emphasis

# Create the pie chart
fig, ax = plt.subplots(figsize=(10, 8))
wedges, texts, autotexts = ax.pie(
    sizes, 
    explode=explode, 
    labels=labels, 
    colors=colors,
    autopct='%1.1f%%',
    shadow=True, 
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)

# Customize text properties
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

for text in texts:
    text.set_fontsize(12)

# Add title and additional information
plt.title('Average Visitor to Order Conversion Rate', fontsize=16, pad=20)

# Add annotation with actual numbers
plt.annotate(
    f'Average Monthly Visitors: {avg_visitors:,.0f}\nAverage Monthly Orders: {avg_orders:,.0f}',
    xy=(0, -0.1),
    xycoords='axes fraction',
    ha='center',
    va='center',
    fontsize=12,
    bbox=dict(boxstyle="round,pad=0.5", facecolor='#f0f0f0', alpha=0.8)
)

# Ensure the pie chart is drawn as a circle
plt.axis('equal')

plt.tight_layout()

# Save the plot as PNG
plt.savefig(os.path.join(charts_dir, 'conversion_rate_pie_chart.png'), dpi=300, bbox_inches='tight')
plt.show()
```
